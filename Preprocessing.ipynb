```python
import pandas as pd

data2 = pd.read_csv("Data/cover_letters.csv",header=1)
```

```python
data2.head()
```

```python
result = []

Q = ""
A = ""
for company,position,val in zip(data2['Organization Name'],data2['Role'],data2['Content']) : 
    for text in val.split("\n") : 
        if not text : continue # 비어있는 값 제외
        if '"' == text : continue # " 로만 되어잇는 데이터 제외
        if text == " " : continue # 공백데이터 제외

        try : 
            # 시작하는 단어가 1,2,3,4로 구분 + 다음글자가 .인지 구분
            if(0<=ord(text[0])-48<=9 and text[1]=='.') :
                if A : result[-1].append(A) # 대답값이 있을때는 이전 배열에 저장
                Q = text[text.index('.')+1:] # 첫 번째 .으로 구분
                result.append([company,position,Q]) # 회사,직군,질문 넣기
                A = ""
            else : 
                A+=text # 대답 쌓기
        except : 
            # 아니면 전부버리기(당장 쓸필요 X)
            pass

result[-1].append(A) # 대답 넣기
```

```python
df = pd.DataFrame(result)
df = df.dropna(axis=0) # 결측치 데이터 제거
df = df[df[2].str.strip() != ''] # 질문에 공백데이터 제거
df = df[df[3].str.len() < 2000] # 2천자 넘어가면 질문 1,2,3,4가 묶여있다고 판단
df = df[df[3].str.len() > 200] # 200자 아닌 데이터는 삭제 
df.to_csv("Data/conver_letters(Fix).csv", index=False, encoding="utf-8-sig")
```
