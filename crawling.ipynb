```python
import requests
from bs4 import BeautifulSoup
import json

jobs = []
for find_text in ['frontend','backend','프론트엔드','백엔드'] : 
# for find_text in ['frontend'] : 
    page=1
    
    while True : 
        # 잡코리아 프론트엔드 채용 공고 URL
        URL = f"https://www.jobkorea.co.kr/Search/?stext={find_text}&careerType=1&tabType=recruit&Page_No={page}"
        # User-Agent 설정 (봇 차단 방지를 위해 필요할 수 있음)
        HEADERS = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
        }

        # 요청 보내기
        response = requests.get(URL, headers=HEADERS)

        # 응답 확인
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            # 채용 공고 정보 추출
            empty_list = soup.find_all("p","list-empty-result")
            if len(empty_list)>=2 : break
            job_list = soup.find_all("article", class_="list-item")
            for job in job_list:
                try : 
                    jobname = json.loads(job['data-gainfo'])['dimension44']
                except : 
                    break
                title_tag = job.find("a", class_="information-title-link dev-view")
                company_tag = job.find("a", class_="corp-name-link dev-view")
                link = title_tag["href"] if title_tag else ""
                title = title_tag.text.strip() if title_tag else "No Title"
                company = company_tag.text.strip() if company_tag else "No Company"
                if title == 'No Title' : continue
                jobs.append({"title": title,"occupation":jobname, "company": company, "link": f"https://www.jobkorea.co.kr{link}"})
            
        page+=1
```
